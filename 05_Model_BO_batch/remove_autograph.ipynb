```python
import os
current_directory = os.getcwd()

print(current_directory)
```

```python
import os
import shutil
directory = '__pycache__'

# Check if the directory exists
if os.path.exists(directory):
    # Remove the directory and all its contents
    shutil.rmtree(directory)
    print(f"Directory {directory} has been removed.")
else:
    print(f"Directory {directory} does not exist.")
```

```python
import glob

# Pattern to match the files
pattern = "__autograph_generated_file*"

# Full path pattern
full_path_pattern = os.path.join(current_directory, pattern)
print(full_path_pattern)

# Find and delete the files
for file in glob.glob(full_path_pattern):
    print(f"Deleting file: {file}")
    os.remove(file)

print("Deletion complete.")
```
